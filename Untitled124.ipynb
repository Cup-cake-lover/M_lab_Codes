```python
#import scienceplots
import numpy as np
import matplotlib.pyplot as plt ; import matplotlib
from numba import jit
import pandas as pd
matplotlib.rcParams['figure.dpi']=100
kernel = np.array([[1,1,0],[1,0,1],[0,1,1]])
#plt.style.use(['science','grid','no-latex'])

@jit(nopython=True)
def init_lattice(L):
  '''
  args : L = lattice size (L)
  returns : ndarray shape L
  '''
  lattice = 2 * np.random.randint(0, 2, L) - 1 ##arrange [-1,+1] spins accordingly
  return lattice

lattice = init_lattice(5)
```

```python
@jit(nopython=True)
def create_target_params(N):
  a = np.random.uniform(0,1,size=(N,N))
  for i in range(N):
    for j in range(N):
      if i==j:a[i,j] = 0


  J = np.tril(a) + np.tril(a, -1).T
  h = np.random.uniform(0,1,size=N)

  return J,h

@jit(nopython=True)
def numba_dot_product(a, b):
    result = 0.0
    for i in range(a.shape[0]):
        result += a[i] * b[i]
    return result

@jit(nopython=True)

def numba_column_stack(arrays):
    rows = arrays[0].shape[0]
    cols = len(arrays)
    result = np.zeros((rows, cols), arrays[0].dtype)

    for i in range(cols):
        result[:, i] = arrays[i]

    return result

@jit(nopython=True)
def custom_mean(arr):
    rows, cols = arr.shape
    result = np.zeros(rows, dtype=np.float64)

    for i in range(rows):
        for j in range(cols):
            result[i] += arr[i, j]

        result[i] /= cols

    return result

@jit(nopython=True)
def fill_diagonal(arr, value):
    rows, cols = arr.shape
    for i in range(min(rows, cols)):
        arr[i, i] = value
    return arr

@jit(nopython=True)
def state_split(states):
  new_state = []
  for i in states:
    for j in i:
      new_state.append(j)

  return new_state
```

```python
@jit(nopython=True)
def calculate_energy_diff(i,lattice,J,h):
  configuration = lattice.copy()
  E = 2*configuration[i]*(numba_dot_product(J[:,i],configuration) + h[i])
  return E

@jit(nopython=True)
def metropolis_hastings_dynamics(lattice,J,h):
  step = 0 ; steps = len(lattice) ; L = len(lattice)
  configuration = lattice.copy()
  L = len(configuration)
  while step < steps:
    i = np.random.randint(0,L)
    delta_E = calculate_energy_diff(i,configuration,J,h)
    if np.random.uniform(0,1) < np.exp(-delta_E):
      configuration[i] *= -1
    step += 1

  return configuration
```

```python
@jit(nopython=True)
def custom_mean(arr):
    rows, cols = arr.shape
    result = np.zeros(rows, dtype=np.float64)

    for i in range(rows):
        for j in range(cols):
            result[i] += arr[i, j]

        result[i] /= cols

    return result

@jit(nopython=True)
def mcmc_sampler(lattice,J,h,M,burn_in):
  sweep = 0 ; _ = 0
  configuration = lattice.copy()
  L = len(configuration)
  C_ij = np.zeros((L,L)) ; states = []

  while _ < burn_in:
    configuration = metropolis_hastings_dynamics(configuration,J,h)
    _ += 1

  for sample in range(5*M):
    if sample % 5 == 0:
      configuration = metropolis_hastings_dynamics(configuration,J,h)
      C_ij += np.outer(configuration,configuration)
      states.append(configuration)

  Mags = custom_mean(numba_column_stack(states))
  C_ij /= M

  return Mags,C_ij,states
```

```python
n = np.arange(17)
Markov_sweeps = 2**n
L = len(lattice)
M = np.zeros((len(n), L ))
C = np.zeros((len(n), L**2))
burn_in = 0

J,h = create_target_params(5)
for i in range(len(n)):
  m,c,_ = mcmc_sampler(lattice,J,h,Markov_sweeps[i],burn_in)
  M[i,:] = m
  C[i,:] = c.flatten()
```

```python
@jit(nopython=True)
def update_params(M_theta,M_d,Corr_theta,Corr_d,eta,J,h):
  h_inf = h + eta*(M_d - M_theta)
  J_inf = J + eta*(Corr_d - Corr_theta)
  J_inf = fill_diagonal(J_inf,0)

  return J_inf,h_inf
```

```python
from collections import Counter
@jit(nopython=True)
def numba_counter(states):
    counts_dict = dict()

    for state in states:
        if state in counts_dict:
            counts_dict[state] += 1
        else:
            counts_dict[state] = 1

    unique_states = np.array(list(counts_dict.keys()))
    counts = np.array(list(counts_dict.values()))

    return counts

@jit(nopython=True)
def negative_log_likelihood(states):
    N = len(states)

    counts = numba_counter(states)
    PMF = counts / N
    NLL = -(1/N) * np.sum(counts * np.log(PMF))

    return NLL
```

```python
@jit(nopython=True)
def boltzmann_learning_machine(lattice_size,sweeps,Mags_d,Corrs_d,iterations,eta,progress_proxy):
  N = lattice_size ; NLL = np.zeros(iterations)
  J,h = create_target_params(N)
  burn_in = 2**17
  for iter in prange(iterations):
    Mags_theta,Corrs_theta,states = mcmc_sampler(init_lattice(N),J,h,sweeps,burn_in)
    J,h = update_params(Mags_theta,Mags_d,Corrs_theta,Corrs_d,eta,J,h)
    print(J,h)
    NLL[iter] = negative_log_likelihood(state_split(states))
    progress_proxy.update(1)

  return J,h,NLL
```
